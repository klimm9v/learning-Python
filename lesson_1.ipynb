```python
# В питоне все является объектом и каждый объект принадлежит какому-либо классу, например
print(type(1)) # <class 'int'>
print(type('2')) # <class 'str'>
print(type(1.1)) # <class 'float'>
print(type([1, 2, 3])) # <class 'list'>
print(type({1, 2, 3})) # <class 'set'>
print(type({1: 'one'})) # <class 'dict'>
print(type((1, 2, 3))) # <class 'tuple'>
print(type(min)) # <class 'builtin_function_or_method'>

# В Python уже сущесвуют встроенные классы, но программисты могут создавать свои классы.
```

```python
# Идея создания классов в программировании была взята из жизни. Например объект яблоко относится к классу растений. Яблоко имеет определенные "свойства" (сорт, цвет, размер, вкус)
# и с яблоком можно взаимодействовать "методами" (откусить, порезать, положить). Каждый объект имеет методы и свойства.
# для создание класса нужно написать ключевое слово class и имя класса, в конце поставить двоеточие.

class ИмяКласса:
    pass

#Внешне классы похожи на функции, но есть и свои различия, например название класса принято писать в верблюжием стиле" (CamelStyle). 
# то есть все слова слитно и начинаются с большой буквы.
```

```python
# Атрибуты класса. Давайте создадим класс Apple и добавим к нему атрибуты (свойства)

class Apple: # определение класса
    form = 'round' # атрибут класса Apple
    color = 'green' # атрибут класса Apple
    taste = 'sweet' # атрибут класса Apple

# form, color, taste - атбрибуты класса Apple. Атрибуты класса похожи на переменные и создаются таким же образом с помощью оператора присваивания '='.
# В примере мы создали 3 атрибута и передали им значения. Допустимые имена атрибутов такие же как и у переменных.
```

```python
# Обращение  к атрибутам класса.
# После того как мы создали атрибуты класса к ним надо как то обратится, прямое обращение атрибутов делается через точку, оператор точка используется для доступа к атрибутам и методам.

print(Apple.form) # вызов атрибута класса Apple, выведет - round
print(Apple.color) # вызов атрибута класса Apple, выведет - green
print(Apple.taste) # вызов атрибута класса Apple, выведет - sweet
```
