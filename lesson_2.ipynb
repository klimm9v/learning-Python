```python
# Экземпляры класса

""" С помощью класса мы можем создавать объекты которые называются - экземпляры. Экземпляры это как "дети классов." 
Каждый экземпляр наследует возможности класса.Например имеет доступ к атрибутам класса или может запускать методы классов (о них позже) 
Зачем нужны экземпляры? В практике программирования, часто работают с экземплярами класса, а не с классом напрямую. 
Класс один, а экземпляров можно создать огромное множество и каждый экземпляр будет наследовать возможонсти класса, при этом может иметь и свои 
уникальные данные. Это как в жизни, дети похожи на родителей, но имеют свои уникальные особенности. """
```

```python
# Создание экземпляра класса
# для создания экземпляра используется конструкция, пример:
class Apple:
    color = 'green'

apple1 = Apple()
apple2 = Apple()
apple3 = Apple()

""" В данном коде мы создали три экземпляра (apple_1, apple_2, apple_3). Каждый экземпляр, при создании, получает доступ к атрибутам класса, то есть может
обращаться к ним напрямую. Чтобы обратиться к атрибуту через экземпляр используется та же конструкция что и в классах, через точку """

print(apple1.color) # green
print(apple2.color) # green
print(apple3.color) # green
```
